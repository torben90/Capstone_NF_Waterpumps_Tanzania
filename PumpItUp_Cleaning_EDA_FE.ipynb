### Plotting Function

```python
def plot_distribution(dataset, cols=5, width=20, height=45, hspace=0.5, wspace=0.5):
    plt.style.use('seaborn-whitegrid')
    fig = plt.figure(figsize=(width,height))
    fig.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=wspace, hspace=hspace)
    rows = math.ceil(float(dataset.shape[1]) / cols)
    for i, column in enumerate(dataset.columns):
        ax = fig.add_subplot(rows, cols, i + 1)
        ax.set_title(column)
        if dataset.dtypes[column] == np.object:
            g = sns.countplot(y=column, data=dataset)
            substrings = [s.get_text()[:18] for s in g.get_yticklabels()]
            g.set(yticklabels=substrings)
            plt.xticks(rotation=25)
        else:
            g = sns.distplot(dataset[column],kde=False) # kde=False if problems, but then the distribution is missing
            plt.xticks(rotation=25)
```

### Cleaning Functions

```python
# Lets see, which columns have empty values
# Therefore a little function which replaces nonsense values with nans

def replace_zero_with_nan(feat_lst, df_lst):
    '''This function helps to change zeros in a column into np.nan inplace
    Input_variables:
    - feat_list: list of features to be transformed
    - df_lst: list of dataframes for which this should take place. Needs to be a list even for only one colums
    Return:
    - None: the transformation will be done inplace so this function has no return
    '''
    for df in df_lst:
        for feat in feat_lst:
            df[feat].replace(0.0, np.nan, inplace=True)
        df['latitude'].values[df['latitude'].values > -0.5] = np.nan
        df['wpt_name'].replace('none', np.nan, inplace=True)
        
        
def fillna_obj(feat_list, df_lst):
    '''This function helps to fill missing values in object-type columns into 'unknown'
    Input_variables:
    - feat_list: list of features to be transformed
    - df_lst: list of dataframes for which this should take place. Needs to be a list even for only one colums
    '''
    for df in df_lst:
        for feat in feat_list:
            df[feat].fillna('unknown', inplace=True)
            

def fillna_bool(feat_list, df_lst, to=True):
    '''This function helps to fill missing values in bool-type columns into True
    Input_variables:
    - feat_list: list of features to be transformed
    - df_lst: list of dataframes for which this should take place. Needs to be a list even for only one colums
    - to: Value after transformation. Default: True
    '''
    for df in df_lst:
        for feat in feat_list:
            df[feat].fillna(to,inplace=True)
            print(feat,'> missing after transformation: ',df[feat].isna().sum()) 
            

def location_nans(feat_lst, df_lst):
    '''This function helps to handle missing values within the location features inplace
    Input_variables:
    - feat_list: list of features to be taken care of
    - df_lst: list of dataframes for which this should take place. Needs to be a list even for only one colums
    Return:
    - No return as the missing value handling will be inplace
    '''
    for df in df_lst:
        for feat in feat_lst:
            df[feat].fillna(df.groupby('subvillage')[feat].transform("median"), inplace=True)
            df[feat].fillna(df.groupby('lga')[feat].transform("median"), inplace=True)
            df[feat].fillna(df.groupby(['region','district_code'])[feat].transform("mean"), inplace=True)
            df[feat].fillna(df.groupby('region')[feat].transform("mean"), inplace=True)
            print(feat,': missing values: ',df[feat].isna().sum())
            
```

## time_in_operation

Create a time between construction year and date_recorded to get a better understanding of the age of a pump

```python
df_time = df.copy()
```

```python
df_time.construction_year.describe()
```

```python
df_time.date_recorded.describe()
```

```python
df_time.head()
```

```python
def operation_time(df_lst):
    for df in df_lst:
        df.date_recorded = pd.to_datetime(df['date_recorded'])
        df['year_recorded'] = df.date_recorded.dt.year
        df['time_in_operation']=df.year_recorded-df.construction_year
        df.drop(['date_recorded','construction_year','year_recorded'],1,inplace=True)
        df['time_in_operation'] = df['time_in_operation'].apply(lambda x: x if x > 0 else df.time_in_operation.mean())
```

```python
operation_time([df_time])
df_time.head()
```

## Grouping feature value with only a small amount of counts

I will now have a look into the futures with a bunch of unique values to see if we can group these to be able to gain some informations out of these features.
* installer
* funder

```python
df_group = df.copy()
```

```python
def group_feat(df_lst):
    for df in df_lst:
        for i in ['funder','installer']:
            df[i] = df[i].str.lower()
            df.loc[(df[i] == 'danid'),i] = 'danida'
            df.loc[(df[i] == '0'),i] = 'unknown'
            df.loc[(df[i] == 'commu'),i] = 'community'
            for c in ['gover','central government','district council',
                      'government of tanzania','gove','lga','governme','central govt']:
                df.loc[(df[i] == c),i] = 'government'
            for c in ['rc church', 'kkkt', 'dmdd','tcrs']:
                df.loc[(df[i] == c),i] = 'religion'    
            for c in ['ces','norad', 'danida','world vision','jaica','unicef','oxfam','amref','rc','wedeco'
                      ,'red cross','japan','holland','germany','germany_republi','netherlands']:
                df.loc[(df[i] == c),i] = 'dev_aid_int'
            for c in ['hesawa','tasaf','rwssp','twesa' ]:
                df.loc[(df[i] == c),i] = 'ngo_nat'

            vc = df[i].value_counts()
            mask = df[i].isin(vc.index[vc<500])
            df.loc[mask, i] = 'other'
```

```python
group_feat([df_group])
```

## Defining a function for the whole feature engineering

```python
def feature_eng(df_lst):
    group_feat(df_lst)
    update_region(df_lst)
    operation_time(df_lst)
```
